```python
import pandas as pd

# Function to find min and max values of any column
def find_min_max(df, column_name):
    min_value = df[column_name].min()
    max_value = df[column_name].max()
    return min_value, max_value

# Load the dataset
df = pd.read_csv("Car_sales.csv")

# Now call the function to find min and max for 'Horsepower'
hp_min, hp_max= find_min_max(df, "Horsepower")

# Print results
print("\nHorsepower - Min:", hp_min)
print("\nHorsepower - Max:", hp_max)
```
